```python
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt

df = pd.read_csv("../data/eth_level1_data.csv")
```

```python
def plot_histogram(column_name):
  ax = df[column_name].hist()
  ax.set_xlabel(column_name)
  ax.set_ylabel("Number of Occurances")
  ax.set_title(f"Histogram of {column_name}, Ethereum 2019-2022")

def plot_vs_price(column_name):
  plt.scatter(x=df[column_name], y=df['Price'], alpha=0.05)
  plt.xlabel(column_name)
  plt.ylabel("Price")
  plt.title(f"Price vs {column_name}, Ethereum 2019-2022")

print(df.columns)
  
```

```python
plot_vs_price("Market Cap")
```
